### Exercise 1

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
training_df = pd.read_csv("Credit_card.csv")
training_df
```

```python
def encode_data(feature_name):
    mapping_dict = {}
    unique_values = list(training_df[feature_name].unique())
    for idx in range(len(unique_values)):
        mapping_dict [unique_values[idx]] = idx
    return mapping_dict

X = training_df[['GENDER', 'Car_Owner','Propert_Owner', 'CHILDREN', 'Annual_income', 'Type_Income','EDUCATION', 'Marital_status',
                 'Housing_type', 'Employed_days', 'Type_Occupation', 'Family_Members']]
y = training_df['label']

X['Type_Income'].replace(encode_data('Type_Income'),inplace = True)
X['EDUCATION'].replace(encode_data('EDUCATION') , inplace = True)
X['Marital_status'].replace(encode_data('Marital_status') ,inplace= True)
X['Housing_type'].replace(encode_data('Housing_type'),inplace = True)
X['Type_Occupation'].replace(encode_data('Type_Occupation'),inplace= True)
X['GENDER'] = X['GENDER'].apply(lambda x: 1 if x == 'M' else 0)
X['Car_Owner'] = X['Car_Owner'].apply(lambda x: 1 if x == 'Y' else 0)
X['Propert_Owner'] = X['Propert_Owner'].apply(lambda x: 1 if x == 'N' else 0)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.callbacks import TensorBoard
import datetime

class MyModel(tf.keras.Model):
    def __init__(self,num_classes=2):
        super(MyModel, self).__init__()
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = Dense(128, activation='relu')
        self.dense2 = Dense(num_classes, activation='softmax')

    def call(self, inputs):
        x = self.flatten(inputs)
        x = self.dense1(x)
        return self.dense2(x)

    # Create an instance of the model
model = MyModel()
```

```python
#Define input Layer using the Functional API
input_layer = Input(shape=(12,))
output_layer = model(input_layer)
```

```python
# Create the model using the Functional API
functional_model = tf.keras.Model(inputs=input_layer,outputs=output_layer)

# Compile the model
functional_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=0)
print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of training label:', y_test.shape)
```

```python
log_dir = "/" + "tensorresults"
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

functional_model.fit(X_train, y_train, epochs=5, callbacks=[tensorboard_callback])
```

```python
training_df2 = pd.read_csv("customer_profile.csv")
training_df3 = pd.read_csv("income_profile.csv")

training_df3['Type_Occupation'] = training_df3['Type_Occupation'].fillna("Not Defined")
training_df3['Annual_income'] = training_df3['Annual_income'].fillna(0)
```

```python
from sklearn.model_selection import train_test_split

def encode_data(feature_name):
    mapping_dict = {}
    unique_values = list(training_df2[feature_name].unique())
    for idx in range(len(unique_values)):
        mapping_dict[unique_values[idx]] = idx
    return mapping_dict

def encode_data2(feature_name):
    mapping_dict = {}
    unique_values = list(training_df3[feature_name].unique())
    for idx in range(len(unique_values)):
        mapping_dict[unique_values[idx]] = idx
    return mapping_dict
```

```python
X_customer = training_df2[['GENDER','Car_Owner','Propert_Owner','CHILDREN','EDUCATION','Marital_status',
                'Housing_type','Employed_days','Family_Members']]

X_income = training_df3[['Annual_income','Type_Income','Type_Occupation']]

y_approve = training_df3['approve']
y_priority = training_df3['priority']

X_customer['EDUCATION'].replace(encode_data('EDUCATION'),inplace = True)
X_customer['Marital_status'].replace(encode_data('Marital_status'),inplace = True)
X_customer['Housing_type'].replace(encode_data('Housing_type'),inplace = True)
X_customer['GENDER'] = X_customer['GENDER'].apply(lambda x: 1 if x == 'M' else 0)
X_customer['Car_Owner'] = X_customer['Car_Owner'].apply(lambda x: 1 if x == 'Y' else 0)
X_customer['Propert_Owner'] = X_customer['Propert_Owner'].apply(lambda x: 1 if x == 'N' else 0)

X_income['Type_Income'].replace(encode_data2('Type_Income'),inplace = True)
X_income['Type_Occupation'].replace(encode_data2('Type_Occupation'),inplace = True)
```

```python
X_income_train, X_income_test, X_customer_train, X_customer_test, y_approve_train, y_approve_test, y_priority_train, y_priority_test = train_test_split( X_income, X_customer, y_approve, y_priority, test_size=0.2, random_state=42)
```

```python
X_customer
```

```python
X_income
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.callbacks import TensorBoard
import datetime
from tensorflow.keras.models import Model

input_income = Input(shape=(X_income_train.shape[1],), name='input_income')
input_customer = Input(shape=(X_customer_train.shape[1],), name='input_customer')

shared_dense = Dense(64, activation='relu')(Concatenate()([input_income, input_customer]))

output_approve = Dense(1, activation='sigmoid', name='output_approve')(shared_dense)

output_priority = Dense(1, activation='sigmoid', name='output_priority')(shared_dense)

model2 = Model(inputs=[input_income, input_customer], outputs=[output_approve, output_priority])
```

```python
model2.compile(optimizer='adam',
               loss={'output_approve': 'binary_crossentropy', 'output_priority': 'binary_crossentropy'},
               metrics={'output_approve': 'accuracy', 'output_priority': 'accuracy'})

model2.summary()
```

```python
log_dir = "/" + "boardresult"
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)


model2.fit({'input_income': X_income_train, 'input_customer': X_customer_train},
           {'output_approve': y_approve_train, 'output_priority': y_priority_train},
           epochs=5,
           validation_data=({'input_income': X_income_test, 'input_customer': X_customer_test},
                            {'output_approve': y_approve_test, 'output_priority': y_priority_test}),
           callbacks=[tensorboard_callback])
```
